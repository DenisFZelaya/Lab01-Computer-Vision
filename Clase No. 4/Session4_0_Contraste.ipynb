```python
import cv2 as cv
```

```python
img01 = cv.imread("flowers.jpeg")
img02 = cv.imread("flowers.jpeg", cv.IMREAD_GRAYSCALE)
```

```python
from matplotlib import pyplot as plt
plt.imshow(cv.cvtColor(img01, cv.COLOR_BGR2RGB))
plt.title("imagen")
plt.show()

```

```python

# Crear funcion para mostrar img
def imshow(img = None, title = ""):
    plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
imshow(img02, "img02")
```

```python
import numpy as np

# Crea un arreglo y asigna por defecto el valor 100
M = np.ones(img02.shape, dtype ="uint8") * 50
```

```python
M
```

Operaciones basicas entre imágenes

```python
img_suma = np.clip(img02.astype(np.int16) + M, 0, 255).astype(np.uint8)
```

```python
imshow(img_suma, "Suma")
```

```python
img_resta = np.clip(img02.astype(np.uint16) - M, 0, 255).astype(np.uint8)
```

```python
imshow(img_resta, "img_resta")
```

```python
constante_M = np.ones_like(img01) * 100
print(img01[67,28])
print(constante_M[67,28])
img_suma_color = np.clip(img01.astype(np.int16) + constante_M, 0, 255).astype(np.uint8)
imshow(img_suma_color)
```

```python
## Equalizar por colores
B, G, R = cv.split(img01)

# Definir constantes a sumar a cada canal
c_azul = 20
c_verde = 30
c_rojo = 100

canal_azul, canal_verde, canal_rojo = cv.split(img01)

c_azul_const = np.clip(canal_azul.astype(int) + c_azul, 0, 255).astype(np.uint8)
c_verde_const = np.clip(canal_verde.astype(int) + c_verde, 0, 255).astype(np.uint8)
c_rojo_const = np.clip(canal_rojo.astype(int) + c_rojo, 0, 255).astype(np.uint8)

img_final = cv.merge((c_azul_const,c_verde_const, c_rojo_const ))

imshow(img_final, "Final")
```
